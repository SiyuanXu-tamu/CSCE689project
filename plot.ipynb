```python
import json
import numpy as np


q_learning_score_path = 'scores/'
sarsa_score_path = 'SARSA_score/'
x = [i for i in range(250, 10050, 250)]

q_max_scores = []
q_avg_scores = []
for i in range(250, 10050, 250):
    path = q_learning_score_path+ 'scores_'+str(i)+'.json'
    with open(path, 'r') as fw:
        injson = json.load(fw)
    score_list = [int(j) for j in list(injson.keys())]
    number_list = list(injson.values())    
    max_score = np.max(score_list)
    test_sum = np.sum(number_list)
    avg_score = 0.0
    for j in range(len(score_list)):
        avg_score += 1.0*number_list[j]/test_sum*score_list[j]
    q_max_scores.append(max_score)
    q_avg_scores.append(avg_score)

s_max_scores = []
s_avg_scores = []
for i in range(250, 10050, 250):
    path = sarsa_score_path+ 'scores_'+str(i)+'.json'
    with open(path, 'r') as fw:
        injson = json.load(fw)
    score_list = [int(j) for j in list(injson.keys())]
    number_list = list(injson.values())    
    max_score = np.max(score_list)
    test_sum = np.sum(number_list)
    avg_score = 0.0
    for j in range(len(score_list)):
        avg_score += 1.0*number_list[j]/test_sum*score_list[j]
    s_max_scores.append(max_score)
    s_avg_scores.append(avg_score)
    

import matplotlib.pyplot as plt

fig=plt.figure(figsize=(8,8),dpi=150)
#plt.axis('equal')
plt.plot(x,q_max_scores,c='blue',label='max_scores')
plt.plot(x,q_avg_scores,c='red',label='avg_scores')
plt.legend(['max_scores','avg_scores'],loc = (0.75,0.9))
plt.xlabel('epoch',fontsize=18)
plt.ylabel('score',fontsize=18)
name = 'Qlearning_scores'
plt.savefig(name+'.png', bbox_inches='tight')
plt.show()

fig=plt.figure(figsize=(8,8),dpi=150)
#plt.axis('equal')
plt.plot(x,s_max_scores,c='blue',label='max_scores')
plt.plot(x,s_avg_scores,c='red',label='avg_scores')
plt.legend(['max_scores','avg_scores'],loc = (0.75,0.9))
plt.xlabel('epoch',fontsize=18)
plt.ylabel('score',fontsize=18)
name = 'SARSA_scores'
plt.savefig(name+'.png', bbox_inches='tight')
plt.show()


fig=plt.figure(figsize=(8,8),dpi=150)
#plt.axis('equal')
plt.plot(x,q_max_scores,c='blue',label='Q-learning_max')
plt.plot(x,q_avg_scores,c='red',label='Q-learning_avg')
plt.plot(x,s_max_scores,c='green',label='SARSA_max')
plt.plot(x,s_avg_scores,c='black',label='SARSA_avg')
plt.legend(['Q-learning_max','Q-learning_avg', 'SARSA_max', 'SARSA_avg'],loc = (0.75,0.9))
plt.xlabel('epoch',fontsize=18)
plt.ylabel('score',fontsize=18)
name = 'SARSA_VS_Q-learning'
plt.savefig(name+'.png', bbox_inches='tight')
plt.show()
```
